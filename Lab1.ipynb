```python
import random
import time
import matplotlib.pyplot as plt


def search_in_sequence(sequence, value):
    return value in sequence


def generate_random_sequence(length):
    seq = list(range(1, length + 1))
    random.shuffle(seq)
    return seq


def plot_time_complexity(x, y_start, y_middle, y_end):
    plt.plot(x, y_start, marker='o', label='Start (Best Case)')
    plt.plot(x, y_middle, marker='o', label='Middle (Average Case)')
    plt.plot(x, y_end, marker='o', label='End (Worst Case)')
    plt.xlabel('Sequence Length (n)')
    plt.ylabel('Average Time (seconds)')
    plt.title('Linear Search Time Complexity Analysis')
    plt.legend()
    plt.grid(True)
    plt.show()


sequence_lengths = [5000, 10000, 15000, 20000]
num_trials = 100


average_times = {'Start': [], 'Middle': [], 'End': []}


for length in sequence_lengths:
    for position in ['Start', 'Middle', 'End']:
        total_time = 0
        for _ in range(num_trials):
            sequence = generate_random_sequence(length)

            # Choose the value based on position
            if position == 'Start':
                value = sequence[0]
            elif position == 'Middle':
                value = sequence[length // 2]
            elif position == 'End':
                value = sequence[-1]

            start_time = time.time()
            search_in_sequence(sequence, value)
            end_time = time.time()

            total_time += (end_time - start_time)
        average_time = total_time / num_trials
        average_times[position].append(average_time)


plot_time_complexity(sequence_lengths, average_times['Start'], average_times['Middle'], average_times['End'])

```
